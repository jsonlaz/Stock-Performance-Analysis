```python
tickers = ["AAPL", "MSFT", "GOOGL", "TSLA", "AMZN", "NVDA", "META"]

stocks_data = yf.download(tickers, start="2020-01-01", end="2024-12-27")

stocks_data
```

```python
stocks_data.to_csv("multi_stock_data.csv")
```

```python
daily_returns = stocks_data['Close'].pct_change().dropna()

daily_returns
```

```python
# Descriptive Analysis
```

```python
daily_returns.describe()
```

```python
# Correlation Analysis
```

```python
correlation_matrix = daily_returns.corr()

correlation_matrix
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np
```

```python
sns.heatmap(daily_returns.corr(), annot = True)

plt.show
```

```python
# Volatility Analysis
```

```python
volatility = daily_returns.std()

volatility
```

```python
annualized_volatility = daily_returns.rolling(window=252).std() * (252**0.5) #annualized volatility
```

```python
annual_return = (1 + daily_returns.mean())**252 - 1

annual_return
```

```python
# Cumulative Returns
```

```python
cumulative_returns = (1 + daily_returns).cumprod()

cumulative_returns
```

```python
# Save as CSV
```

```python
stocks_data['Date'] = stocks_data.index
daily_returns['Date'] = daily_returns.index
cumulative_returns['Date'] = cumulative_returns.index
annualized_volatility['Date'] = annualized_volatility.index
```

```python
stocks_data.to_csv('stock_data.csv')
daily_returns.to_csv('daily_returns.csv')
cumulative_returns.to_csv('cumulative_returns.csv')
annualized_volatility.to_csv('volatility.csv')
```
